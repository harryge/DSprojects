## Percentage of total not fraud transactions

```python
nfcount=0
notFraud=data_df['Class']
for i in range(len(notFraud)):
  if notFraud[i]==0:
    nfcount=nfcount+1
nfcount
per_nf=(nfcount/len(notFraud))*100
print("Percentage of total not fraud transaction in the dataset",per_nf)
```

## Percentage of total fraud transactions

```python
nfcount=0
Fraud=data_df['Class']
for i in range(len(Fraud)):
  if Fraud[i]==1:
    nfcount=nfcount+1
nfcount
per_nf=(nfcount/len(Fraud))*100
print("Percentage of total fraud transaction in the dataset",per_nf)
```
